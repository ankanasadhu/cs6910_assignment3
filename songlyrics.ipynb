```python
gpt2.download_gpt2(model_name="124M")
```

```python
file_name = "/content/train.txt"
```

```python
sess = gpt2.start_tf_sess()
gpt2.finetune(sess,
          steps=600,
          dataset=file_name,
          model_name='124M', 
          run_name= 'run1',
          restore_from='latest',
          print_every=100,
          sample_every=100,
          save_every=100
          )
```

```python
gpt2.generate(sess,
              length=25,
              temperature=0.7,
              prefix="I love deep learning",
              nsamples=2,
              batch_size=2
              )
```
